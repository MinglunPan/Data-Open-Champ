```python
df_groupby = df.groupby(['year','effylev'])[GRAND_TOTAL,GRAND_TOTAL_MAN,GRAND_TOTAL_WOMAN].sum().reset_index()
years = np.unique(df_groupby['year'])
import matplotlib.pyplot as plt
fig,ax = plt.subplots()
width = 0.13
w = 0
ind = np.arange(len(years)) + .15 
for i,n in enumerate([GRAND_TOTAL,GRAND_TOTAL_MAN,GRAND_TOTAL_WOMAN]):
    df_all = df_groupby[df_groupby.effylev == effylev_breakdown[0]][n]
    df_undergrad = df_groupby[df_groupby.effylev == effylev_breakdown[1]][n]
    df_grad = df_groupby[df_groupby.effylev == effylev_breakdown[2]][n]
    rect1 = ax.bar(ind+w, df_all,width)
    rect2 = ax.bar(ind+w, df_undergrad,width )
    rect3 = ax.bar(ind+w, df_grad,width )
    w += 0.15
    
extra_space = 0.05
ax.set_xticks(ind+width+extra_space)
ax.set_xticklabels(years)
```

```python
for i in effylev_breakdown:
    _df = df_groupby[df_groupby['effylev'] == i]
    _df.plot(kind = 'bar',x = 'year',y = [GRAND_TOTAL,GRAND_TOTAL_MAN,GRAND_TOTAL_WOMAN],title = i)
    plt.legend(loc='lower left')
```

```python
race_category = [name.lower() for name in np.unique(EFFY_descp['varname']) if name not in list(EFFY_descp.iloc[0:7,1]) and name.endswith('T')]
label_name = [ EFFY_descp.loc[EFFY_descp.varname == name.upper(),'shortDescription'].iloc[0] for name in race_category]
```

```python
for r, l in zip(race_category, label_name):
    print(r,l)
```

```python
df_groupby2 = df.groupby(['year','effylev'])[race_category].sum().reset_index()
```

```python
for i in effylev_breakdown:
    _df = df_groupby2[df_groupby2['effylev'] == i]
    _df.plot(kind = 'bar',x = 'year',y = race_category,title = i,label = label_name)
    plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
```
